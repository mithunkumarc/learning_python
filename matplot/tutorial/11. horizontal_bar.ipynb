```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Fixing random state for reproducibility
np.random.seed (19680801)
# rcdefaults : Restore the rc params from Matplotlib's internal default style.
plt.rcdefaults()
fig, ax = plt.subplots()  #ax : AxesSubplot , subplots() ; Create a figure and a set of subplots.
# example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people)) # y axis values
performance = 3 + 10 * np.random.rand(len(people)) # adding 3 to  random numbers
error = np.random.rand(len(people)) # for tip on the top of bar 
# bar horizontal, xerr : , add horizontal / vertical errorbars to the bar tips. , ecolor : color of error
ax.barh(y_pos, performance, xerr = error, align = 'center',color='green',ecolor='red')
ax.set_yticks(y_pos)
ax.set_yticklabels(people) # text value for y axis
ax.invert_yaxis() # inverting y axis : people appears left to right
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
plt.show()
```

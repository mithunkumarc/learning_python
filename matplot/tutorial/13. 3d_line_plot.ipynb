```python
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
fig = plt.figure()
# GCA : Get the current Axes instance on the current figure matching the given keyword args, or create one.
ax = fig.gca(projection = '3d')#3d plot
theta = np.linspace(-4 * np.pi , 4 * np.pi, 100) # angle
z = np.linspace(-2,2,100) # generate 100 random numbers between 2 and -2
r = z**2 + 1
x = r * np.sin(theta) # gen x values
y = r * np.cos(theta) # gen y values
ax.plot(x,y,z, label = 'parametric curve demo')
ax.legend()
plt.show()
```
